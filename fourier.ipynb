```python
import numpy as np
import matplotlib.pyplot as plt
```

# Convolution

```python
from scipy.stats import norm

SIZE = 400
# axis (In image processing, the unit is pixel)
x = np.array(range(SIZE))

# make ground truth
f = np.where( (x>100) & (x<300),1,0)
plt.show()

# make pdf (as noize)
n = norm.pdf(x,loc = SIZE/2, scale = 10)

# make observed sigmal
g = np.convolve(f,n,mode ='same')

plt.plot(f)
plt.plot(g)
plt.show()

plt.plot(n)
plt.show()
```

# Make fourier transformation of probability density funciton

 This process is needed because np.convolution is not idea and right noize is not n but n_new.

```python
# Considering circular convolution and np.convolve, n_new is right noize from f to g.
n_new = np.concatenate((n,n))[200:600]

plt.plot(n_new)
plt.show()
plt.plot(np.fft.fft(n_new))
plt.show()
```
